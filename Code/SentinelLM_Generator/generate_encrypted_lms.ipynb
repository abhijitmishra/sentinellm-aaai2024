```python
%%writefile upload_script.py
from huggingface_hub import HfApi, Repository
import os

def upload_model(model_path, model_name):
    api = HfApi()
    hf_username = api.whoami(token=os.environ["HF_TOKEN"])["name"]
    repo_id = f"{hf_username}/{model_name}"

    # create repos
    try:
        api.create_repo(token=os.environ["HF_TOKEN"], name=model_name, organization=hf_username, exist_ok=True)
    except Exception as e:
        print(f"Repository creation failed or already exists: {e}")

    # upload folder
    api.upload_folder(
        folder_path=model_path,
        repo_id=repo_id,
        token=os.environ["HF_TOKEN"]
    )
```

```python
%%bash
models="roberta-base bert-base-uncased bert-base-multilingual-uncased"
keys="llm123 nlp2023"
embedding_rescaling_factors="3 10"

for model in $models; do
  for key in $keys; do
    for erf in $embedding_rescaling_factors; do
      MODEL_PATH="${model}-${key}-fact${erf}"
      MODEL_NAME="${model}-${key}-fact${erf}"
      echo "Uploading $MODEL_NAME to Hugging Face Hub"
      python -c "from upload_script import upload_model; upload_model('$MODEL_PATH', '$MODEL_NAME')"
      echo "Upload of $MODEL_NAME completed"
    done
  done
done
```
